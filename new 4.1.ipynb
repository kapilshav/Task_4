```python
#Task_4.2
import cv2
import numpy

#dorae
doraemon = cv2.imread('dorae.jpg')

cv2.imshow('doraemon', doraemon)
cv2.waitKey(0)
cv2.destroyAllWindows()

#shin
shin = cv2.imread('shinchan.jpg')
cv2.imshow('shinchan', shin)
cv2.waitKey(0)
cv2.destroyAllWindows()


#crop_dorae
crop_dorae=doraemon[0:230,100:300]
cv2.imshow('Cropped_dorae', crop_dorae)
cv2.waitKey(0)
cv2.destroyAllWindows()

#crop_shin
crop_shin=shin[0:230,100:300]
cv2.imshow('cropped_shin', crop_shin)
cv2.waitKey(0)
cv2.destroyAllWindows()

#SWapping


crop_hun = numpy.copy(crop_dorae)
crop_hun2 = numpy.copy(crop_shin)
doraemon[0:230,100:300]=crop_hun2
shin[0:230,100:300]= crop_hun



doraemon[0:230 , 100:300]= crop_hun2
cv2.imshow(('dorae_shin'), doraemon)
cv2.waitKey(0)
cv2.destroyAllWindows()


shin[0:230 , 100:300]= crop_hun
cv2.imshow(('shin_dorae'),shin)
cv2.waitKey(0)
cv2.destroyAllWindows()






```
